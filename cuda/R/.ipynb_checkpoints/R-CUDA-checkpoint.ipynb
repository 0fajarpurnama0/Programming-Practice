Matrix multiplication in R using CPU.

```r
library(gputools)
matA <- matrix(2,7000)
matB <- matrix(3,1,7000)

# Start the clock!
ptm <- proc.time()

cpuMatMult(matA, matB)

proc.time() - ptm
```

Matrix multiplication in R using GPU.

```r
library(gmatrix)

host_a <- runif(100,1,10)
host_b <- runif(100,1,10)

#Creating GPU objects
gpu_a = g(host_a)
gpu_b = g(host_b)

# Start the clock!
ptm <- proc.time()

#Perform some simple calculations
gpu_c = gpu_a + gpu_b

# Stop the clock!
proc.time() - ptm

#Move data back to the host
host_c = h(gpu_c)
```
